```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(100)
data = np.random.normal(9000,3000,300)
data = data[data>=4000]
data
```

```python
len(data)
```

```python
plt.hist(data,color='r',rwidth=0.7,alpha=0.6)
plt.show()
```

```python
plt.hist(data,bins=20)
plt.show()
```

```python
len(data[(data>7000) & (data<9500)])
```

```python
np.random.seed(100)
da = np.random.normal(size=100)
plt.hist(da)
plt.show()
```

```python
da = np.concatenate([da,[4,7,9,-4,-8]])
```

```python
plt.boxplot(da)
plt.show()
```

```python
plt.boxplot(da,whis=3)
plt.show()
```

```python
plt.boxplot(da,showmeans=True,meanline=True,showbox=False,widths=1)
plt.show()
```
