```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
x
```

```python
np.random.seed(1)
y = np.random.random(size=10)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='r',linestyle='--',marker='o')  # - -- -. :
plt.show()
```

```python
plt.plot(x,y,'b--')
plt.show()
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
y2 = np.cos(x)
y2
```

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

```python
plt.plot(x,y,label='sin(x)')
plt.plot(x,y2,label='cos(x)')

plt.xlabel('x axis')
plt.ylabel('y value')
plt.title('sin cos')
plt.legend()
plt.show()
```

```python
plt.plot(x,y,label='sin(x)')
plt.plot(x,y2,label='cos(x)')

plt.xlabel('x axis')
plt.ylabel('y value')
plt.title('sin cos')
plt.legend() #表示绘制图例
plt.axis([-4,12,-2,2])  #横坐标-4到12  纵坐标-2到2
plt.show()
```
