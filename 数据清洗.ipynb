```python
import pandas as pd
```

```python
data={
    'name':['djf', 'jij', 'ief', None, 'lij'],
    'age':[18, None, 32 ,25, None],
    'score':['A', 'B', None, 'B', None]
}
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.dropna()  # dropna丢弃null数据
```

```python
df.dropna(how='all') # 丢弃所有是null的数据
```

```python
df.dropna(subset=['age']) # 丢弃age中的null数据
```

```python
df.fillna(0)
```

```python
df.fillna({'age':18,'score':'D'})
```

```python
df
```

```python
df.fillna(method='ffill') #根据缺失值前面的值进行填充
```

```python
df.fillna(method='bfill') #根据缺失值后面的值进行填充
```

```python
df.fillna({'age':df.age.mean()})
```

```python
df.fillna({'age':df.age.median()})
```

```python
# 插值
df
```

```python
df.interpolate()
```

```python
df.fillna({'score':'missing'},limit=1)
```

```python
df.fillna({'score':'missing'},limit=1,inplace=True)
```

```python
df
```

## 异常值处理

```python
df.loc[5]={'name':'test', 'age':999, 'score':'C'}
```

```python
df
```

```python
q_upper = df['age'].quantile(0.75) #上4分位数
```

```python
q_lower = df['age'].quantile(0.25) #下4分位数
```

```python
k = 1.5
```

```python
d = df[df['age']>q_upper+k*(q_upper-q_lower)] # 取出age列大于  上4分位数+k乘以上-下
```

```python
d
```

```python
df
```

```python
df.drop(d.index)
```

```python
df[(df['age'] < q_upper+k*(q_upper-q_lower)) & (df['age'] > q_lower-k*(q_upper-q_lower))]
```
